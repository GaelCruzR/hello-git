## Cargar Librerias

```python
# Importar librerias necesarias para los DF
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Cargar Dataset

```python
df = pd.read_csv("WA_Fn-UseC_-Marketing-Customer-Value-Analysis.csv")

df.head()
```

### Configuración general del estilo

```python
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)
```

## Gráficas

### Distribución de clientes por tipo de canal de ventas (Gráfica de Barras)

```python
sales_counts = (
    df["Sales Channel"].value_counts().sort_values(ascending=False)
)

plt.bar(sales_counts.index, sales_counts.values)
plt.title("El canal de ventas por agente concentra la mayor cantidad de clientes")
plt.xlabel("Canal de ventas")
plt.ylabel("Número de clientes")
plt.tight_layout()
plt.show()
```

### Distribución del Customer Lifetime Value (CLV) (Histograma)

```python
plt.hist(
    df["Customer Lifetime Value"],
    bins=30
)
plt.title("La mayoría de los clientes presenta un valor de vida medio")
plt.xlabel("Customer Lifetime Value")
plt.ylabel("Frecuencia")
plt.tight_layout()
plt.show()
```

### Comparación del Total Claim Amount por tamaño de vehículo (Boxplot)

```python
order = ["Small", "Medsize", "Large"]

sns.boxplot(
    data=df,
    x="Vehicle Size",
    y="Total Claim Amount",
    order=order
)

plt.title("Los vehículos grandes presentan mayor variabilidad en reclamaciones")
plt.xlabel("Tamaño del vehículo")
plt.ylabel("Monto total de reclamaciones")
plt.tight_layout()
plt.show()
```
